```python
import pandas as pd

data = pd.read_csv("result.csv")

data
```

```python
grouped = data.groupby(['Buff_size', 'Prod_or_cons', 'PC_config', 'Is_fair', 'Randomization'])

grouped.count()
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,5)


def draw_plot(data):
    x = []
    y = []
    for i, j in data.items():
        x.append(str(i))
        y.append(j)
        
    plt.scatter(x, y)
    plt.xlabel('Portion size')
    plt.ylabel('Buffer access number')
    plt.show()

    
def get_iterator():
    for grouped_by, grouped_data in grouped:
        print(f"Buffer size: {grouped_by[0]}, Worker: {grouped_by[1]}, Worker number: {grouped_by[2]}, Is fair buffer: {grouped_by[3]}, Randomization type: {grouped_by[4]}")
        data_to_show = grouped_data.groupby("Size")["Buff_access_num"].sum()
        
        data_to_show.plot.box()
        print(plt.show())
        print()
        print(draw_plot(data_to_show))
        
iterator = get_iterator()
```
