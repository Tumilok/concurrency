```python
import pandas as pd

data = pd.read_csv("result.csv")

data
```

```python
grouped = data.groupby(['Buff_size', 'Prod_or_cons', 'PC_config', 'Is_fair', 'Randomization'])

grouped.count()
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,5)

def draw_plot(data):
    x = []
    y = []
    for i,j in data.items():
        x.append(str(i))
        y.append(j)
    plt.scatter(x,y)
    plt.show()
    
def f():
    for group,data_for_group in grouped:
        print(group)
        print(f"Buffer size is {group[0]}, {group[1]}, number of {group[1]}s is {group[2]}, normal randomization: {group[4]}, fair buffer: {group[3]}")
        print(group)
        a = data_for_group.groupby("Size")["Buff_access_num"].sum()
        a.plot.box()
        yield plt.show()
        print()
        
        yield draw_plot(a)
```

```python
iterator = f()
next(iterator)
next(iterator)
next(iterator)
next(iterator)
```
